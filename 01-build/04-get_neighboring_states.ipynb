```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/ubikuity/List-of-neighboring-states-for-each-US-state/master/neighbors-states.csv'
neighbors = pd.read_csv(url)
neighbors.head()
```

```python
neighbors = neighbors[neighbors.NeighborStateCode != 'FL'] # we don't have Florida data
```

```python
neighbors['state'] = neighbors['StateCode'].map(abbrev)
neighbors['neigh_state'] = neighbors['NeighborStateCode'].map(abbrev)
```

```python
# missing states
missing_states = pd.Series([i for i in abbrev.keys() if i not in neighbors.StateCode.unique()]).map(abbrev)
missing_states
```

```python
neighbors = neighbors.groupby('state')[['neigh_state']].agg(lambda x: list(x)).reset_index()
```

```python
# add missing states
names = ['Hawaii', 'Rhode Island', 'South Carolina', 'Texas', 'Vermont', 'Washington', 'West Virginia', 'Wisconsin', 'Wyoming']
neighs = [['California'],
          ['Massachusetts', 'Connecticut'],
          ['North Carolina', 'Georgia', ],
          ['Louisiana', 'Oklahoma', 'New Mexico', 'Arkansas'],
          ['Massachusetts', 'New York', 'New Hampshire'],
          ['Maryland', 'Virginia'],
          ['Ohio', 'Virginia', 'Kentucky', 'Maryland', 'Pennsylvania'],
          ['Michigan', 'Minnesota', 'Iowa', 'Illinois'],
          ['Montana', 'Idaho', 'Utah', 'Colorado', 'Nebraska', 'South Dakota']]
new_states = pd.DataFrame()
new_states['state'] = names
new_states['neigh_state'] = neighs
neighbors = neighbors.append(new_states).set_index('state')
```
